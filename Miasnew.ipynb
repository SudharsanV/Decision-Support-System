```
from google.colab import files
uploaded =files.upload()
import numpy as np 
from PIL import Image
trainM=[]
for filename in uploaded.keys():
  if filename.endswith('.png'):   
    with Image.open(filename) as image:
      trainM.append((np.array(image)))
y_l2=list()
for i in range(0,len(trainM)):
  y_l2.append(0)
          
```

```
x_label=trainB+trainM
y_label=y_l+y_l2
#shuffledata
import random
c=list(zip(x_label,y_label))
random.shuffle(c)
x_image,y_image=zip(*c)
print(y_image)
```

```
import cv2
y_imag=np.array(y_image)
resizedimage=list()

for image in x_image:
  re=cv2.resize(image,(300,300))
  resizedimage.append(re)

resized_image=np.array(resizedimage)
for i in range (0,len(resized_image)):
  print(resized_image[i].shape)
  
  
resized_image=resized_image.reshape((resized_image.shape[0], resized_image.shape[1], resized_image.shape[2],1))
resized_image.shape

```

```
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Convolution2D, MaxPooling2D
from keras.models import Model
from keras.layers import Activation, Dropout, Flatten, Dense
```

```
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense

model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape=(300, 300, 1)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))


# the model so far outputs 3D feature maps (height, width, features)

model.add(Flatten())  # this converts our 3D feature maps to 1D feature vectors
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
model.add(Activation('sigmoid'))
# COMPILE
model.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
```

```
k=0
count=0
tp=0
tn=0
fp=0
fn=0

y_val=list()
z=len(resized_image) 
print(z)
for i in range(0,z-1):
  if(i==0):
    a=resized_image[i]
    y_hat=y_imag[i]
    y_imag=np.delete(y_imag,i)
    resized_image=np.delete(resized_image,(i),axis=0)
    a=a.reshape((a.shape[2],a.shape[1], a.shape[0],1)) 
    model.fit(resized_image,y_imag,epochs=10)
    ans=model.predict_proba(a)
    b=a
    print(ans)
    
  else:
   
    a=resized_image[i]
    y_hat=y_imag[i]
    y_imag=np.delete(y_imag,i)
    resized_image=np.delete(resized_image,(i),axis=0)
    b=b.reshape((b.shape[0],b.shape[1],b.shape[2],1))    
    a=a.reshape((a.shape[2],a.shape[1],a.shape[0],1))
    resized_image=np.append(resized_image,a,axis=0)
    y_imag=np.append(y_imag,y_hat)
    
 
  resized_image.shape
  model.fit(resized_image,y_imag,epochs=5)
  predicted=model.predict_proba(a)
  print(predicted[0][0])
  if(predicted[0][0]>=0.5):
    ans=1
  else:
    ans=0
  print(ans)
  print(y_hat)
  
  if((y_hat==1) and (ans==1)):
    tp=tp+1
  elif((y_hat==0) and (ans==0)):
    tn=tn+1
  elif((y_hat==0) and (ans==1)):
    fp=fp+1
  else:
    fn=fn+1      
      
```

```
accuracy=((tp+tn)/(tp+fp+tn+fn))
precision=(tp/(tp+fp))
recall=(tp/(tp+fn))
fmeasure=((2*precision*recall)/(precision+recall))
tpr=recall
fpr=(fp/(fp+fn))
tnr=(tn/(tn+fp))
fnr=(fn/(tp+fn))
```

```
print("Accuracy" , accuracy*100)
print("Precision" , precision)
print("Recall" , recall)
print("F-measure" , fmeasure)
print("True Positive ", tp)
print("True Negative ", tn)
print("False Positive ", fp)
print("False Negative ", fn)
print("True Positive Rate ", tpr)
print("False Positive Rate ", fpr)
print("True Negative Rate ", tnr)
print("False Positive Rate ", fnr)
model.summary()
 
```
